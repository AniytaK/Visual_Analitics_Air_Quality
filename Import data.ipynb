##Step 1. Read and merge historical data for "Main data set".


```python
!pip install fastparquet

import pandas as pd
import glob

file_directory = "D:/Studium/3_HS 2023/Visual Analitics/Ideen/DE/ParquetFiles/E1a/"

file_pattern = "*.parquet"
```

```python
dfs = []

for file in glob.glob(file_directory + file_pattern):
    df = pd.read_parquet(file)
    dfs.append(df)
combined_df = pd.concat(dfs, axis=0) 
```

```python
from datetime import datetime
from meteostat import Point, Daily
import matplotlib.pyplot as plt

#z.B. fÃ¼r Berlin, Jahr 2018
start = datetime(2018, 1, 1)
end = datetime(2018, 12, 31)
berlin = Point(52.5200, 13.4050, 70)
data = Daily(berlin, start, end)
data = data.fetch()

data.plot(y=['tavg', 'tmin', 'tmax'])
plt.show()
```
